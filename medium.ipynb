```python
# LC_Course Schedule

# 想不到

# 主要知識點：
# This problem is equivalent of finding if a cycle exists in a directed graph. use DFS or topological sorting. 
# The solution is really brilliant, using dfs!
# Time Complexity: O(V+E). (because of adjacency list)

class Solution(object):
    def canFinish(self, numCourses, prerequisites):
        graph = [[] for _ in range(numCourses)] # in order to convert prerequisites to adjancency list representation of graph
        visit = [0 for _ in range(numCourses)]
        for x, y in prerequisites:
            graph[x].append(y)
            
        def dfs(i):
            if visit[i] == -1:
                return False
            if visit[i] == 1:
                return True
            visit[i] = -1
            for j in graph[i]:
                if not dfs(j):
                    return False
            visit[i] = 1
            return True 
    
        for i in range(numCourses):
            if not dfs(i):
                return False
        return True
    
# The solution below uses topological sorting (implemented with BFS, i.e. kahn's algorithm). 
# The topological sort is natural for this problem. We always take the courses with no unstudied prereqs
# and so on until no more courses we can take. The oud[i] is the number of prereqs for course i and indegree 
# keep a list of courses require course i.
# Time complexity: O(V+E), look at the code comments below
# Space complexity: O(V+E), oud+ind

def canFinish(self, n, pres):
    from collections import deque
    ind = [[] for _ in xrange(n)]  # indegree
    oud = [0] * n  # outdegree
    
    # O(E)
    for p in pres:
        oud[p[0]] += 1
        ind[p[1]].append(p[0])
    dq = deque()
    # O(V)
    for i in xrange(n):
        if oud[i] == 0:
            dq.append(i)
    k = 0
    # I'm not sure but I think the piece of code below is O(V+E), for we will have V times of ind[x], and we 
    # will have run through each edge in the end (for i in ind[x]), which is O(E). 
    while dq:
        x = dq.popleft()
        k += 1
        for i in ind[x]:
            oud[i] -= 1
            if oud[i] == 0:
                dq.append(i)
    return k == n
```

```python
# LC_Clone Graph

# 想不到

# 主要知識點：
# DFS. the stack here makes the code DFS.
# return a deep copy: use nodeHash to do the trick. 

class Solution(object):
    def cloneGraph(self, node):
        if not node:
            return node
        
        nodeHash = {}
        nodeHash[node] = Node(node.val, [])
        stack = [node]
        while stack:
            currNode = stack.pop()
            for neighNode in currNode.neighbors:
                if not neighNode in nodeHash:
                    nodeHash[neighNode] = Node(neighNode.val, [])
                    stack.append(neighNode)
                nodeHash[currNode].neighbors.append(nodeHash[neighNode])
                
        return nodeHash[node]
```

```python
# LC_Simplify Path

# 想不到

# 主要知識點：
# The trick behind the solution, or programming: break your final solution into pieces
# For this example, path_noslash = deque([i for i in path.split("/") if i not in ["", "."]])  is a gooood first step
# Time Complexity: O(N)
# Space Complexity: O(N)

from collections import deque

class Solution(object):
    def simplifyPath(self, path):
        path_noslash = deque([i for i in path.split("/") if i not in ["", "."]])
        l = [""]
        while path_noslash:
            pos = path_noslash.popleft()
            if pos == ".." :
                if  l[-1] != "":
                    l.pop()
            else:
                l.append(pos)
                
        return "/".join(l) if len(l) > 1  else "/"
```

```python
# LC_Design Circular Queue

# 想不到

# 主要知識點：
# One thing to note here about the problem discription: once the queue becomes full, we cannot insert the next 
# element even if there is a space in front of the queue... is that it assumes that queue is implemented with arrays not
# linked list. a linked list queue can grow or shrink as much as needed. 
# if we use arrays to implement queue (linear), we use two pointers (referrence) to the index
# to keep track of the front and the rear.

class MyCircularQueue(object):

    def __init__(self, k):
        self.q = [None] * k
        self.len = k
        self.front = 0
        self.rear = 0
        
    def enQueue(self, value):
        if self.q[self.rear] is None:
            self.q[self.rear] = value
            self.rear = (self.rear+1) % self.len # the % self.len trick makes the queue circular
            return True
        else:
            return False

    def deQueue(self):
        if self.q[self.front] is None:
            return False
        else:
            self.q[self.front] = None
            self.front = (self.front+1) % self.len
            return True
        
    def Front(self):
        return -1 if self.q[self.front] is None else self.q[self.front]

    def Rear(self):
        return -1 if self.q[self.rear - 1] is None else self.q[self.rear - 1]
        
    def isEmpty(self):
        return self.front == self.rear and self.q[self.front] is None

    def isFull(self):
        return self.front == self.rear and self.q[self.front] is not None
```

```python
# LC_Task Scheduler

# 想不到 (有多種解法)

# 主要知識點：
# The solution is so impressive! it basically approaches from math perspective. (math is so important )
# Time Complexity: O(n) (We iterate over tasks array only once, to get Counter. sorted(c.values) is constant! because
# c = Counter(tasks), tasks ranges from A to Z, which means c is length 26 tops.)
# Space Complexity: O(1) (c = Counter(tasks), tasks ranges from A to Z)

from collections import Counter
class Solution(object):
    def leastInterval(self, tasks, n):        
        c = Counter(tasks)
        vals = sorted(c.values(), reverse = True)
        
        max_val = vals[0]-1
        max_idle_slots = max_val*n
        vals.pop(0)
        
        for val in vals:
            max_idle_slots -= min(max_val, val)
        return len(tasks) + max_idle_slots if max_idle_slots > 0 else len(tasks)
    
    
# below is a priority queue solution.

# 主要知識點:
# the key in the approach here is that we always prioritize the most frequent signals, to make total time minimum.
# Thus, priority queue is what we can use.
# heapq in python always pop the smallest one. In our scenario, we have to add negative sign to the counts
# When slots become zero (slots filled with other signals), we stop while loop and start over again.
# This time, the priroty queue, again, pops the smallest. 
# When there is no element in heap of the while slots and heap, this means that we can't find any signal to fill in the 
# slots, we must then use idle. and time += idle.
# Time Complexity: O(n)  (Counter) (don't bother with while loop. while loop won't exceed c*n, where c is a constant)
# Space Complexity: O(1) (c = Counter(tasks), tasks ranges from A to Z)

from collections import Counter
import heapq

class Solution(object):
    def leastInterval(self, tasks, n):
        time = 0
        counter_tasks = Counter(tasks)
        heap = []
        for task in counter_tasks:
            heapq.heappush(heap, (-counter_tasks[task], task))
            
        while heap:
            slots = n+1
            temp_ls = []
            while slots and heap:
                count_, task = heapq.heappop(heap)
                
                if count_ < -1:
                    temp_ls.append((count_+1, task))
                time += 1
                slots -= 1
            
            for i in temp_ls:
                heapq.heappush(heap, i)
                
            if heap:
                time += slots
        
        return time 
```

# DP (Dynamic Programming)

```python
# LC_Coin Change (最少硬幣組合)

# 想不到

# 主要知識點：
# 解法用Dynamic programming - Bottom up
# Time complexity : O(S*n), where S是amount, n是len(coins)
# On each step the algorithm finds the next F(i) in n iterations, S*n
# Space complexity : O(S)
# We use extra space for the memoization table. (where the concept of dp comes in.)

class Solution(object):
    def coinChange(self, coins, amount):
        """
        :type coins: List[int]
        :type amount: int
        :rtype: int
        """
        rs = [amount+1] * (amount+1)
        rs[0] = 0
        for i in range(1, amount+1):
            for c in coins:
                if i >= c:
                    rs[i] = min(rs[i], rs[i-c] + 1)

        if rs[amount] == amount+1:
            return -1 # 這個if如果成立，就代表..amount無法被coins組成
        return rs[amount]
    
    
x = Solution()
x.coinChange([1, 2, 5], 11)
```

```python
# LC_Product of Array Except Self

# 想不到

# 主要知識點
# god-like solution

def productofTheOthers(arr):
    
    s = len(arr)
    r = []
    
    k = 1 
    for i in range(s):
        r += [k]
        k *= arr[i]
        
    k = 1
    for i in reversed(range(s)):
        r[i] *= k 
        k *= arr[i]
        
    return r
print(productofTheOthers([3,9,7,-2]))
```
